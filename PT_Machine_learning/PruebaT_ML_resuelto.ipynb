### Ejercicio 3.

```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
# URL de la página de comercio electrónico (Ejemplo)
url = 'https://www.onlybabys.com.ar/otono-invierno-2024/remeras1/'
```

```python
# Realizar la solicitud HTTP a la página web
result = requests.get(url).text
```

```python
# Crear una variable con el nombre soup  y lo traducimos con el parser lxml
# Usamos la función BeautifulSoup, pasamos como argumento el texto extraido y el parser
soup = BeautifulSoup(result, 'lxml')
```

```python
# Encontrar los contenedores de productos (divs con clase específica en este caso)
product_list = soup.find_all('div', class_ = 'js-item-container item')
```

```python
# Busca el nombre del producto en el documento HTML que se encuentra en un <div> con la clase  'js-item-name item-name mt-1 mb-3 font-small opacity-80'
# Luego, extrae el texto dentro de ese elemento.
product_name = soup.find('div', class_= 'js-item-name item-name mt-1 mb-3 font-small opacity-80').text

# Busca el precio del producto en el documento HTML en un elemento <span> con la clase específica 'js-installment-price product-installment-value'
# Luego, extrae el texto dentro de ese elemento.
product_price = soup.find('span', class_ ='js-installment-price product-installment-value').text

# Imprime el nombre del producto y precio
print(product_name)
print(product_price)
```

```python
# Inicializa una lista vacía donde se almacenarán los productos extraídos.
products_onlybaby= []

# Itera sobre cada elemento en la lista de productos obtenida anteriormente.
for lista_productos in product_list:
    # Busca el nombre del producto y el precio dentro del elemento actual de la lista de productos.
    # Luego, extrae el texto de los elementos y elimina los espacios en blanco alrededor.
    name  = lista_productos.find('div', class_= 'js-item-name item-name mt-1 mb-3 font-small opacity-80').text.strip() 
    price = lista_productos.find('span', class_='js-installment-price product-installment-value').text.strip() 
    
# Agrega un diccionario con el nombre y el precio del producto a la lista de productos.
    products_onlybaby.append({'title': name, 'price': price})
    
# Imprime la lista de los nombres de los productos y sus precios
    print(name)
    print(price)
```

```python
# Guardar los datos extraídos en un archivo JSON
with open('products_onlybaby.json', 'w') as json_file:
    json.dump(products_onlybaby, json_file, indent=4)

print(" Los datos fueron guardados en el archivo (products_onlybaby.json)")
```

### Ejercicio 5.

El modelo One-Class SVM (Support Vector Machine) es una técnica de aprendizaje no supervisado utilizada principalmente para la detección de anomalías. Fue desarrollado inicialmente para clasificación binaria, pero su variante One-Class se adapta para identificar si un nuevo dato pertenece al mismo grupo que el conjunto de entrenamiento o si es una anomalía.

El modelo One-Class SVM se utiliza en diversas áreas, tales como:

* Detección de Fraude: Identificar transacciones fraudulentas en sistemas financieros.
* Mantenimiento Predictivo: Detectar fallos en maquinaria y equipos industriales.
* Seguridad de Redes: Identificar intrusiones o comportamientos anómalos en redes informáticas.
* Calidad de Datos: Identificar errores y valores atípicos en bases de datos.
* Salud: Identificar anomalías en datos médicos que puedan indicar condiciones de salud inusuales.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import OneClassSVM

# Generar datos normales
X_normal, _ = make_blobs(n_samples=300, centers=1, cluster_std=0.5, random_state=42)

# Agregar algunos puntos anómalos
np.random.seed(42)
X_atipico = np.random.uniform(low=1, high=10, size=(20, 2))

# Combina los datos normales con los anómalos
X_data = np.concatenate([X_normal, X_atipico], axis=0)
dataX = X_data.reshape(-1,1)
```

```python
# Entrenar el modelo One-Class SVM
modelo_svm = OneClassSVM(kernel='rbf', gamma=0.1, nu=0.2)
modelo_svm.fit(dataX)  # entrenar con datos  
```

```python
# Predecicción del modelo
y_predice = modelo_svm.predict(dataX)
```

```python
# Crear un diagrama de caja
plt.boxplot(dataX)
plt.title("Diagrama de Caja con Outliers")
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Cargar el dataset Iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Convertir a DataFrame para una visualización más sencilla
df = pd.DataFrame(data= np.c_[X, y], columns=iris.feature_names + ['target'])
display(df.head())
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Crear y entrenar el modelo Random Forest
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train, y_train)
```

```python
# Hacer predicciones en el conjunto de prueba
y_predic = modelo_rf.predict(X_test)

# Evaluar el rendimiento del modelo
accuracy = accuracy_score(y_test, y_predic)
print(f'Accuracy: {accuracy}')

print('Classification Report:')
print(classification_report(y_test, y_predic, target_names=iris.target_names))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt


# Definir las transformaciones para normalizar los datos
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

# Descargar y cargar el conjunto de datos de entrenamiento
trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=100,
                                          shuffle=True, num_workers=2)

# Descargar y cargar el conjunto de datos de prueba
testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=100,
                                         shuffle=False, num_workers=2)

# Definir las clases del dataset
classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        self.fc1 = nn.Linear(32 * 8 * 8, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 32 * 8 * 8)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Crear una instancia de la red
net = SimpleCNN()
```

```python
import torch.nn.functional as F

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
for epoch in range(10):  # número de épocas
    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # Obtener los inputs
        inputs, labels = data

        # Poner los gradientes a cero
        optimizer.zero_grad()

        # Forward + backward + optimización
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # Imprimir estadísticas
        running_loss += loss.item()
        if i % 100 == 99:  # imprimir cada 100 mini-batches
            print(f'[Epoch {epoch + 1}, Batch {i + 1}] loss: {running_loss / 100:.3f}')
            running_loss = 0.0

print('Finished Training')
```

```python
correct = 0
total = 0
with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f'Accuracy of the network on the 10000 test images: {100 * correct / total}%')
```

```python
# Función para mostrar una imagen
def imshow(img):
    img = img / 2 + 0.5  # desnormalizar
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

# Obtener algunas imágenes de prueba
dataiter = iter(testloader)
images, labels = next(dataiter) # Use next(dataiter) instead of dataiter.next()

# Mostrar imágenes
imshow(torchvision.utils.make_grid(images))

# Imprimir etiquetas reales
print(' '.join(f'{classes[labels[j]]:5s}' for j in range(4)))

# Imprimir predicciones
outputs = net(images)
_, predicted = torch.max(outputs, 1)
print(' '.join(f'{classes[predicted[j]]:5s}' for j in range(4)))
```
